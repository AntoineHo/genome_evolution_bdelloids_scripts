### Pairwise comparisons (t-test correction for unequal variance):

```python
form_P500 = formatted_data[5]
form_P250 = formatted_data[4]
form_P100 = formatted_data[3]
form_P0 = formatted_data[2]
form_D30 = formatted_data[1]
form_H30 = formatted_data[0]

def cohens_d(d1, d2) :
    n1, n2 = len(d1), len(d2)
    s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)
    s = np.sqrt( ((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2) )
    u1, u2 = np.mean(d1), np.mean(d2)
    return (u1 - u2) / s

def hedges_g(d1, d2) :
    n1, n2 = len(d1), len(d2)
    dof = n1 + n2 - 2
    s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)
    u1, u2 = np.mean(d1), np.mean(d2)
    diff_mean = abs(u1 - u2)
    s_pooled_star = np.sqrt( (((n1 - 1) * s1) + ((n2 - 1) * s2)) / dof )
    return diff_mean / s_pooled_star

def glass_d(d1, d2) : # WARNING: d1 MUST BE THE CONTROL GROUP
    u1, u2 = np.mean(d1), np.mean(d2)
    diff_mean = abs(u1 - u2)
    s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)
    glasss_delta = diff_mean / np.sqrt(s1)
    return glasss_delta

# alternative=less: the mean of the distribution underlying the first sample is less 
#                     than the mean of the distribution underlying the second sample.

s, p = ss.ttest_ind(form_P500, form_P0, equal_var=False, alternative="two-sided") # H1 : P500 sign lower than P0; H0 : no diff
print("t-test P500 != P0:\nS = {:.3f}\np = {:.3f}".format(s, p))
print("Cohen's d P500 - P0 = {:.3f}".format(cohens_d(form_P500, form_P0)))
print("Hedge's g P500 - P0 = {:.3f}".format(hedges_g(form_P500, form_P0)))
print("Glass' delta P500 - P0 = {:.3f}".format(glass_d(form_P0, form_P500)))
print("\n----------\n")

s, p = ss.ttest_ind(form_P250, form_P0, equal_var=False, alternative="two-sided") # H1 : P500 sign lower than P0; H0 : no diff
print("t-test P250 != P0:\nS = {:.3f}\np = {:.3f}".format(s, p))
print("Cohen's d P250 - P0 = {:.3f}".format(cohens_d(form_P250, form_P0)))
print("Hedge's g P250 - P0 = {:.3f}".format(hedges_g(form_P250, form_P0)))
print("Glass' delta P250 - P0 = {:.3f}".format(glass_d(form_P0, form_P250)))
print("\n----------\n")

s, p = ss.ttest_ind(form_P100, form_P0, equal_var=False, alternative="two-sided") # H1 : P500 sign lower than P0; H0 : no diff
print("t-test P100 != P0:\nS = {:.3f}\np = {:.3f}".format(s, p))
print("Cohen's d P100 - P0 = {:.3f}".format(cohens_d(form_P100, form_P0)))
print("Hedge's g P100 - P0 = {:.3f}".format(hedges_g(form_P100, form_P0)))
print("Glass' delta P100 - P0 = {:.3f}".format(glass_d(form_P0, form_P100)))
print("\n----------\n")

#s, p = ss.ttest_ind(form_P500, form_H30, equal_var=False, alternative="less") # H1 : P500 sign lower than P0; H0 : no diff
#print("t-test P500 < H30:\nS = {:.3f}\np = {:.3f}".format(s, p))

#print("\n----------\n")

s, p = ss.ttest_ind(form_D30, form_H30, equal_var=False, alternative="two-sided") # H1 : P500 sign lower than P0; H0 : no diff
print("t-test D30 != H30:\nS = {:.3f}\np = {:.3f}".format(s, p))
print("Cohen's d D30 - H30 = {:.3f}".format(cohens_d(form_D30, form_H30)))
print("Hedge's g D30 - H30 = {:.3f}".format(hedges_g(form_D30, form_H30)))
print("Glass' delta D30 - H30 = {:.3f}".format(glass_d(form_H30, form_D30))) # control first in glass' delta
print("\n----------\n")

s, p = ss.ttest_ind(form_D30, form_H30, equal_var=False, alternative="less") # H1 : P500 sign lower than P0; H0 : no diff
print("t-test D30 < H30:\nS = {:.3f}\np = {:.3f}".format(s, p))

print("\n----------\n")

s, p = ss.ttest_ind(form_D30, form_H30, equal_var=False, alternative="greater") # H1 : P500 sign lower than P0; H0 : no diff
print("t-test D30 > H30:\nS = {:.3f}\np = {:.3f}".format(s, p))
```
